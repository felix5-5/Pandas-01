```python
import pandas as pd
```

```python
# Create a function to format the dates:

formatter = lambda x: pd.datetime.strptime(x, "%b %d, %Y")
```

```python
# Convert columns to datetime while importing: 

btc = pd.read_csv("data/Bitcoin Historical Data.csv", parse_dates=["Date"], date_parser=formatter)
```

```python
btc
```

```python
# Convert a column to datetime:

btc["Date"] = pd.to_datetime(btc["Date"], format="%b %d, %Y")
```

```python
# Get the day name of an specific date:

btc.loc[5, "Date"].day_name()
```

```python
# Filtering by dates: 

date_filt = (btc["Date"] >= "Sep 01, 2022")
```

```python
# Set dates as the index: 

btc.set_index(["Date"])
```

```python
# Resampling: 

btc["Date"].resample("Y").mean()

btc.resample("W").agg({"Date":"mean"})
```
