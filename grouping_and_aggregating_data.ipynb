```python
import pandas as pd
```

```python
df = pd.read_csv("data/ds_salaries.csv")
```

```python
df
```

```python
# Grab the median of a column:
df["salary_in_usd"].median()
```

```python
# value_counts() method: return how many times a value appears. 
df["company_location"].value_counts().loc[["BR","US"]]
```

```python
# groupby() function: creates a group for each values in that column.
location_grp = df.groupby(["company_location"])
```

```python
# get_group() method: allows to access an specific group.
df.groupby(["company_location"]).get_group("BR")
```

```python
# Search by an specific column for each group and apply a function: 
location_grp["salary_in_usd"].median() 
```

```python
# Apply a function for specific values inside that group:
location_grp["salary_in_usd"].median().loc[["BR", "US"]]
```

```python
# normalize parameter, show te results in porcentage.  
location_grp["remote_ratio"].value_counts(normalize=True).loc[["BR", "US"]]
```

```python
# agg() methodo allows to apply more than one function to the operation. 
location_grp["salary_in_usd"].agg(["median", "min", "max"]).loc[["BR", "US"]]
```
